Part # 2 : Conditionals

```python
""" برنامج لتصنيف قيمة المبيعات المدخلة بواسطة المستخدم الى ضعيف , متوسط , ممتاز"""
Monthly_Sales= float(input("Monthly Sales"))
if Monthly_Sales < 1000 :
    print("ضعيف")
elif Monthly_Sales >= 1000 and Monthly_Sales <= 5000 :
    print("متوسط")
else: 
    print("ممتاز")
```

```python
""" برنامج لتصنيف مخزون منتج ( يتم ادخال قيمة المخزون بواسطة المستخدم ) : نفذ المخزون اذا كان 0 , كمية منخفضة اقل من 10 
متوفر اذا اكتر من 10 """
Quantity_in_stock= float(input("Product Quantity in stock"))
if Quantity_in_stock == 0 :
    print("نفذ المخزون")
elif Quantity_in_stock <10 :
    print("كمية منخفضة")
else:
    print("متوفر")
```

```python
"""
اكتب برنامج ياخذ مبيعات 3 أشهر ويطبع أعلى شهر مبيعات
تم ادخال القيم بواسطة المستخدم 
استخدمنا float لقبول الارقام التي تحتوي على فواصل عشرية
اولا تم المقارنة بحيث اذا انطبق و تحقق الشرطين And في اول if بالتالي الشهر الاول هو لاعلى مبيعا
اذا لم يتحقق واحد من شرطين ال and في اول if سوف ينتقل لل elif
نفس النظام يتطبق على ال elif فاذا لم يتحقق شرطيها  
تخرج ناتج ال else بحيث تكون هي اعلى قيمة
"""
First_month= float(input("First month sales:"))
Second_month= float(input("Second month sales:"))
Third_month= float(input("Third month sales:"))
if First_month >= Second_month and First_month >= Third_month :
    print("أعلى مبيعات  كانت الشهر لاول: ",First_month)
elif Second_month >= First_month and Second_month >= Third_month :
    print("اعلى مبيعات كانت الشهر الثاني: ",Second_month)
else :
    print("أعلى مبيعات كانت الشهر الثالث : ",Third_month)
```

Part # 3 : LOOPs


```python
""" 
1- قمنا بانشاء list لكل من month , sales
2- استخدمنا loop 
توضبح مثال اول مرة  i=0
هي ضمن ال range >>> سوف يتم طباعة 1 و jan , 2000
لان jan , 2000 >>> i index =0  
3- تم استحداث متغير total ليكون هو المكان الذي يتم اضافة القيم بداخله
"""
Months=["Jan","Feb","Mar","Apr","May","Jue","Jul","Aug","Sep","Oct","Nov","Dec"]
Sales=[2000, 2500, 6000, 500, 4120, 3700, 5000, 3000, 1000, 5200, 1900, 6000]
for i in range(12):
    print(i+1 ,Months[i],"Sales = ",Sales[i])
Total = 0


""" لطباعة مجموع المبيعات الاكبر من 2000"""
for y in Sales:
    if y > 2000 :
        Total += y
print("مجموع المبيعات الاعلى من 2000 =",Total)
```

Part # 4 : Functions

```python
def calculate_revenue(price,quantity):
    Revenue = price * quantity
    return(Revenue)

calculate_revenue(15.1,100.0)
```

```python
def apply_discount(price,discount_percent):
    price_after_disount = price - (price * discount_percent/100)
    return (price_after_disount)

apply_discount(500,10)
```

```python
def classify_performance(Monthly_Sales):
    if Monthly_Sales < 1000 :
        return("ضعيف")
    elif Monthly_Sales >= 1000 and Monthly_Sales <= 5000 :
        return("متوسط")
    else: 
        return("ممتاز")

classify_performance(1000)
```

```python
def average(number_list):
    if len(number_list) == 0:
        return (0)
    else:
        return (sum(number_list)/len(number_list))

average([15,14,13,20])
```
