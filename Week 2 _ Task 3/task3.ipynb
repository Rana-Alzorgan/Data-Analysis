مقدمة Numpy

```python
#Part 1 - 1
import numpy as np
np.random.seed(0)
sales= np.random.randint(100,1001,20)
reshaped_sales=sales.reshape(4,5)
print(sales)
print(reshaped_sales)
```

```python
#Part 1 - 2
np.random.seed(0)
branch1_sales= np.random.randint(500,801,5)
branch2_sales=np.random.randint(700,901,5)
print(branch1_sales,branch2_sales)
total_sales=branch1_sales+branch2_sales
sales_diff=branch2_sales-branch1_sales
print("difference = ",sales_diff)
print("total sales = ",total_sales)
```

```python
#Part 1 - 3
sales_mean=np.mean(sales)
sales_std=np.std(sales)
sales_max=np.max(sales)
sales_min=np.min(sales)
print("sales mean = ",sales_mean)
print("sales standard deviation = ",sales_std)
print("maximum sales = ",sales_max)
print("minimum sales = ",sales_min)
# النتائج كانت متطابقة بين Numpy و الدالة اليدوية
```

```python
#Part 1 - 4
print(reshaped_sales[0,]) #استخراج الاسبوع الاول
print(reshaped_sales[1,]) #استخراج الاسبواع الثاني كامل
print(reshaped_sales[ : ,1:2]) #استخراج اليوم الثاني من كل الاسابيع
print(reshaped_sales[1:3,:]) #استخراج قيم الاسبوع الثاني و الثالث 
```

```python
#Part 4 - 1
survival_percentage  = titanic["survived"].mean()*100
survival_percentage_check = np.mean(titanic["survived"]==1)*100
non_survival_percentage = np.mean(titanic["survived"]==0)*100
print(f"نسبة الركاب الناجين = {survival_percentage:0.2f} %")
```

```python
#Part 4 - 2
survival_percentage_by_gender = titanic.groupby("sex")["survived"].mean()*100
print("نسبة الناجين لكل من الاناث و الذكور موزعة كالتالي")
print(survival_percentage_by_gender)
print("تبين النتائج ان نسبة الناجيات من الاناث أعلى من الناجين من الذكور")
```

```python
#Part 4 - 3
survival_percentage_by_pclass = titanic.groupby("pclass")["survived"].mean()*100
print("نسبة الناجين من الركاب حسب فئة التذكرة")
print(survival_percentage_by_pclass)
print("تبين ان نسبة النجاة لركاب الفئة الاولى اعلى بكثير من نسبة النجاة لركاب الفئة الثالثة")
```

```python
#Part 4 - 4
survivors_avg_age = titanic.groupby("survived")["age"].mean()
print("متوسط اعمار الناجين و غير الناجين")
print(survivors_avg_age )
print("يتضح ان متوسط اعمار الغير ناجين يقارب ال 31 عام و متوسط اعمار الناجين يقارب ال 28 عام")
```

ظهر تحليل البيانات أن نسبة النجاة الإجمالية كانت حوالي 38%، مما يعني أن غالبية الركاب لم يتمكنوا من النجاة خلال الحادث. كما تبين وجود اختلاف واضح في معدلات النجاة حسب فئة الراكب، حيث سجل ركاب الفئة الأولى أعلى فرص للنجاة مقارنة بالفئتين الثانية والثالثة، مما يشير إلى ارتباط مستوى الفئة بفرص البقاء، وقد يكون ذلك مرتبطًا بموقع الركاب وسهولة الوصول إلى قوارب النجاة. كذلك ظهر تأثير الجنس كعامل مهم، حيث كانت نسبة نجاة الإناث أعلى من الذكور، وقد يرتبط ذلك بإجراءات الإخلاء التي أعطت الأولوية للنساء. في المقابل، لم يظهر العمر كعامل مؤثر بشكل واضح، حيث كانت متوسطات أعمار الناجين وغير الناجين متقاربة.
