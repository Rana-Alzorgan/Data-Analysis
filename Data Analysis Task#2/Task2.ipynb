```python
#اكتب دالة calculate_median(numbers) ترجع الوسيط (انتبه لحالة العدد الزوجي من العناصر — لازم تاخذ متوسط العنصرين الوسطيين).

def calculate_median(numbers):
    #هون اتاكدت بالاول انه في مدخلات 
    if len(numbers) == 0:
        return None
    #هون بالبداية رتبت القائمة تصاعديا , بعدين اوجدت كم عدد العناصر عندي بالقائمة , بعدين طلعت موقع القيمة الوسطية بحالة القوائم الفردية
    sorted_numbers = sorted(numbers)
    n = len(numbers)
    middle_index = n // 2
    # هون بلشت ابني المعادله median بحيث لو كانت القائمة زوجية فياخد متوسط القيمتين الوسطيتين
    if n%2 == 0 :
        median = ((sorted_numbers[middle_index] + sorted_numbers[middle_index - 1])/2)
        return (median)
    # هون بحالة القائمة الفردية ف يرجعلي القيمة الوسطية 
    else:
        return(sorted_numbers[middle_index])
    
calculate_median([15,1,4,17,5,0])
calculate_median([19,14,50])
calculate_median([])
calculate_median([130,110,500,200,90,150,120])
```

```python
# اكتب دالة calculate_mode(numbers) ترجع القيمة الأكثر تكرارًا (أو أكثر من قيمة لو تعادلوا بالتكرار).

def calculate_mode(numbers):

# هون انشات متغير لحفظ اعلى مرات تكرار و قائمة احفظ فيها القيمه و القيم الاعلى تكرار
    max_count =0
    mode_value = []
# هون عملت for loop : بتاخد العنصر بتعمله كاونتر مصفر بالبدايه و بعدين حلقه داخلها تمر على باقي العناصر
# الحلقة الداخليه رح تقارن العنصر مع لعناصر الاخرى و كل مره بيحدث تكرار بنزيد 1 على الكاونتر وهكذا
# اذا كان التكرار لها العنصر اكبر من قيمة اعلى تكرار يلي بالمتغير max_count بيتم حفظ هاي القيمه ك max_count و بيتم حفظ العنصر ك mode_value
# في حالة وجود اكتر من عنصر يحمل نفس قيمة ال max_count و بنفس الوقت هاد العنصر غير موجود بال mode_value list بيتم اضافة هاد العنصر لل mode_value list
    for y in numbers :
        count = 0
        for x in numbers :
            if x == y :
                count += 1

        if count > max_count :
            max_count = count
            mode_value = [y]
        elif count == max_count and y not in mode_value :
            mode_value.append(y)

    return (mode_value)


#calculate_mode([1,1,1,0,2,5,7,9,8,7,5,3,2,1])
#calculate_mode([1,1,1,8,9,9,8,5,8,7,9])
#calculate_mode([130,110,500,200,90,150,120])
```

```python
#4.اكتب دالة calculate_variance(numbers) ترجع التباين (متوسط مربعات الفرق عن المتوسط).

def calculate_variance(numbers):

    mean = calculate_mean(numbers)
    n= len(numbers)
    total = 0

    for num in numbers : 
       total += (num - mean)**2

    variance = total/n
    return(variance)

calculate_variance([130,110,500,200,90,150,120])
```

الجزء 2 — الأرباعيات (Quartiles) واكتشاف القيم الشاذة (IQR)

```python
#اكتب دالة calculate_quartiles(numbers) ترتّب القائمة وترجع Q1، Q2 (الوسيط)، وQ3 (فيه أكثر من طريقة حساب للأرباعيات، اختر أبسط طريقة منطقية وطبّقها بثبات).

def calculate_quartiles(numbers):

    numbers = sorted(numbers)
    n = len(numbers)

    # حساب قيمة Q2

    if n%2 == 0 :
        q2 = (numbers[n//2] + numbers[n//2 -1]) /2
        lower_half = numbers[:n//2]
        upper_half = numbers[n//2:]
    else:
        q2 = numbers[n//2]
        lower_half = numbers[:n//2]
        upper_half = numbers[n//2 +1:]

    # حساب قيمة Q1 and Q2

    def median(separated_data):
        m = len (separated_data)
        if m%2 == 0 :
            return ((separated_data [m//2]+ separated_data [m//2 -1]) / 2)
        else:
            return (separated_data [m//2])

    q1 = median(lower_half) 
    q3 = median(upper_half) 
    
    return q1,q2,q3
calculate_quartiles([120, 150, 90, 200, 500, 110, 130])
```

```python
#2.اكتب دالة detect_outliers_iqr(numbers) تستخدم الدالة السابقة، تحسب IQR، وترجع قائمة بكل القيم الشاذة يلي وجدتها حسب القاعدة فوق.

def detect_outliers_iqr (numbers):

    q1,q2,q3 = calculate_quartiles(numbers)
    IQR = q3 - q1
    Maximum = q3 + 1.5*IQR
    Minimum = q1 - 1.5*IQR
    Outliers = []

    for number in numbers:
        if number > Maximum or number < Minimum :
            Outliers.append(number)

    return Outliers

detect_outliers_iqr([120, 150, 90, 200, 500, 110, 130])
```

الجزء 3 — Correlation (الارتباط)

```python
def calculate_correlation(x,y):
    
    if len(x) != len(y) :
        return("Error: The number of values in both lists must be the same.")
    
    mean_x = calculate_mean(x)
    mean_y = calculate_mean(y)
    std_x = calculate_std(x)
    std_y = calculate_std(y)
    covariance = 0
    n = len(x)

    for i in range(n):
         covariance += (x[i] - mean_x) * (y[i] - mean_y)

    covariance = covariance / n

    r = covariance / (std_x * std_y)

    return round(r,4)

x=[100, 150, 200, 250, 300, 350]
y=[500, 620, 810, 900, 1100, 1200]
calculate_correlation(x,y)
        
# العلاقة طردية قوية جدا و هذا يعني انه كل ما زاد عدد الزوار كل ما زادت الايرادات بالتالي يجب على صاحب المتجر جذب عدد زوار اكبر لزيادته ايراداته اليوميه
```

الجزء 4 — مقدمة الاحتمالات

```python
def probability_of_an_event(num_of_favourable_outcomes, total_number_of_events_in_the_sample_space):
    probability = num_of_favourable_outcomes / total_number_of_events_in_the_sample_space
    return(f"probability = {probability * 100:.1f}%")
    
probability_of_an_event(40,200)
```

```python
def conditional_probability(favourable_outcomes, total_condition):
    probability = favourable_outcomes / total_condition
    return (f"P(شراء | مشترك بالنشرة) = {probability * 100:.1f}%")

conditional_probability(90, 300)
```
